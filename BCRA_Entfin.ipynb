# Import del CSV


```python
# leer el csv "COMPLETO.TXT" que esta en la siguiente ruta /content/BCRA_enfin/COMPLETO.TXT

import pandas as pd

#  'COMPLETO.TXT' is a CSV file
try:
    df = pd.read_csv('/content/drive/MyDrive/BCRA_enfin/COMPLETO.TXT', sep='\t', encoding='latin-1',header=None,index_col=False) # Specify the separator and encoding
    print(df.head(100))  # Print the first few rows to verify
except FileNotFoundError:
    print("Error: 'COMPLETO.TXT' not found in the specified path.")
except pd.errors.ParserError:
    print("Error: Could not parse the file. Check the file format and separator.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
# Define the new column names
new_column_names = ['cod_entfin', 'nombre', 'fecha', 'cod_categoria', 'categoria', 'debe', 'haber']

# Check if the number of columns matches the number of new names
if len(new_column_names) == len(df.columns):
    df.columns = new_column_names
    print(df.head()) # Print the first few rows to verify
else:
    print("Error: Number of columns in dataframe does not match the number of new column names.")

df
```

Filtro por el codigo 131745 correspondiente a los creditos hipotecarios uva


```python
# : tomar la tabla "df" y filtrar por la variable cod_categoria == 131745 y crear un dataframe nuevo que se llame "cred_hip_uva"

# Filter the DataFrame
cred_hip_uva = df[df['cod_categoria'] == 131745]

# Print some info
print(cred_hip_uva.head())
print(cred_hip_uva.info())
```

ordeno de mayor a menor


```python
# : ordenar el dataframe cred_hip_uva por la variable "debe" de mayor a menor

cred_hip_uva = cred_hip_uva.sort_values(by='debe', ascending=False)
cred_hip_uva
```

mismo procedimiento que el anterior pero con el activo

```python
# Filter the DataFrame
activo = df[df['cod_categoria'] == 100000]


activo = activo.sort_values(by='debe', ascending=False)
# Print some info

print(activo.info())
activo
```

```python
df_clasificador = pd.read_csv('/content/drive/MyDrive/BCRA_enfin/clasificador_entfin.TXT', sep='\t', encoding='latin-1',index_col=False,header=None) # Specify the separator and encoding

# Define the new column names
new_column_names_2 = ['cod_entfin', 'nombre','categoria_entfin']

# Check if the number of columns matches the number of new names
if len(new_column_names_2) == len(df_clasificador.columns):
    df_clasificador.columns = new_column_names_2
    print(df_clasificador.head()) # Print the first few rows to verify
else:
    print("Error: Number of columns in dataframe does not match the number of new column names.")


print(df_clasificador.info())
df_clasificador
```

merge del clasificador con las tablas de creditos y activo

```python
print(activo.info())
print(df_clasificador.info())
print(cred_hip_uva.info())
print(df_clasificador.info())
```

```python
: agregar la columna  categoria_entfin de "df_clasificador" segun "cod_entfin" al  dataframe "activo" y "cred_hip_uva"

# Merge the 'categoria_entfin' column from 'df_clasificador' into 'activo' and 'cred_hip_uva' based on 'cod_entfin'
activo = pd.merge(activo, df_clasificador, on='cod_entfin', how='left')
cred_hip_uva = pd.merge(cred_hip_uva, df_clasificador, on='cod_entfin', how='left')

# Print the updated DataFrames to verify the merge
print(activo.head())
print(cred_hip_uva.head())
```
